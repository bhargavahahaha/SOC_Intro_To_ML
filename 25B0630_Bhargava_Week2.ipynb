```python
# 5 students × 3 subjects (Math, Science, English)
marks = np.array([
    [78, 82, 85],
    [91, 89, 94],
    [67, 72, 70],
    [88, 90, 86],
    [56, 60, 58]
])

student_names = ["Aarav", "Diya", "Rohan", "Ananya", "Kabir"]
subject_names = ["Math", "Science", "English"]

print("Marks Array:")
print(marks)
```



### Numpy performance - Student Performance Analytics
Write the code with reference to the marks array defined above
1. Print:

   a) Shape of the array

   b) Number of dimensions

   c) Data type
2. Calculate:

   a) Total marks of each student

   b) Average marks of each student

3. Find:

   a) Highest marks in each subject

   b) Lowest marks in each subject

4. Add 5 grace marks to every student using broadcasting.
5. Replace all marks below 70 with 0.
6. Find the topper (student with highest total marks).
7. Create a new array containing students whose average marks are above 80.
8. Count how many marks are greater than 85.


```python
print(marks.shape)

print(marks.ndim)

print(marks.dtype)

total_marks = np.sum(marks, axis=1)
for i in range(len(student_names)):
    print(student_names[i], ":", total_marks[i])

average_marks = np.mean(marks, axis=1)
for i in range(len(student_names)):
    print(student_names[i], ":", average_marks[i])

highest_marks = np.max(marks, axis=0)
for i in range(len(subject_names)):
    print(subject_names[i], ":", highest_marks[i])

lowest_marks = np.min(marks, axis=0)
for i in range(len(subject_names)):
    print(subject_names[i], ":", lowest_marks[i])

updated_marks = marks + 5
print(updated_marks)

new_marks = marks.copy()
new_marks[new_marks < 70] = 0
print(new_marks)

total_marks = np.sum(marks, axis=1)
topper_index = np.argmax(total_marks)
print("Topper:", student_names[topper_index])
print("Total Marks:", total_marks[topper_index])

average_marks = np.mean(marks, axis=1)
result = marks[average_marks > 80]
print(result)

count = np.sum(marks > 85)
print(count)
```


# 3. Matplotlib Basics

Matplotlib is used for plotting and visualizing data.


```python
# Note: Make sure you have completed the Pandas section above
# so that the Total_Marks column exists in the dataframe.
# Line plot of student averages

plt.figure(figsize=(8, 5))

plt.plot(df["Name"], df["Average"], marker='o')

plt.title("Student Average Scores")
plt.xlabel("Student Name")
plt.ylabel("Average Score")

plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```

```python

# Bar chart

subject_means = [
    df["Math"].mean(),
    df["Science"].mean(),
    df["English"].mean()
]

subjects = ["Math", "Science", "English"]

plt.bar(subjects, subject_means)

plt.title("Average Marks by Subject")
plt.ylabel("Average Marks")

plt.show()
```

```python

# Histogram

all_marks = pd.concat([
    df["Math"],
    df["Science"],
    df["English"]
])

plt.hist(all_marks, bins=10)

plt.title("Distribution of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("final_assignment_dataset.csv")

print(df.head())
print(df.info())
print(df.describe())

df['Total_Marks'] = df['Math'] + df['Science'] + df['English']
df['Average_Marks'] = df['Total_Marks'] / 3
highest_student = df[df['Average_Marks'] == df['Average_Marks'].max()]
print("\nHighest Scoring Student:")
print(highest_student[['Name', 'Average_Marks']])
lowest_student = df[df['Average_Marks'] == df['Average_Marks'].min()]
print("\nLowest Scoring Student:")
print(lowest_student[['Name', 'Average_Marks']])
above_80 = df[df['Average_Marks'] > 80]
print("\nStudents Scoring Above 80 Average:")
print(above_80[['Name', 'Average_Marks']])

math_array = np.array(df['Math'])
science_array = np.array(df['Science'])
english_array = np.array(df['English'])
print("\nMath - Mean:", np.mean(math_array))
print("Math - Median:", np.median(math_array))
print("Math - Standard Deviation:", np.std(math_array))
print("\nScience - Mean:", np.mean(science_array))
print("Science - Median:", np.median(science_array))
print("Science - Standard Deviation:", np.std(science_array))
print("\nEnglish - Mean:", np.mean(english_array))
print("English - Median:", np.median(english_array))
print("English - Standard Deviation:", np.std(english_array))
subject_means = {
    'Math': np.mean(math_array),
    'Science': np.mean(science_array),
    'English': np.mean(english_array)
}
best_subject = max(subject_means, key=subject_means.get)
print("\nSubject with Highest Average Marks:", best_subject)

plt.figure(figsize=(12, 5))
plt.plot(df['Name'], df['Average_Marks'], marker='o', color='blue', linestyle='-')
plt.title("Line Plot of Student Averages")
plt.xlabel("Student Name")
plt.ylabel("Average Marks")
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 5))
plt.hist(df['Average_Marks'], bins=6, color='green', edgecolor='black')
plt.title("Histogram of Average Marks")
plt.xlabel("Average Marks")
plt.ylabel("Number of Students")
plt.grid(axis='y')
plt.tight_layout()
plt.show()

subject_names = ['Math', 'Science', 'English']
subject_averages = [np.mean(math_array), np.mean(science_array), np.mean(english_array)]

plt.figure(figsize=(7, 5))
plt.bar(subject_names, subject_averages, color=['steelblue', 'orange', 'green'], edgecolor='black')
plt.title("Bar Chart of Subject-wise Averages")
plt.xlabel("Subject")
plt.ylabel("Average Marks")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
